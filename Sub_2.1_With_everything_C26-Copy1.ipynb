Now we create an adjacency matrix to create the financial network to work with.

Rows and columns correspond to the input and output country-industry respectively.

```python
import numpy as np

nodes = [c + " " + i for c in country.keys() for i in industry.keys()]
data = data.loc[nodes][nodes]

adjacency_matrix = np.zeros([len(nodes),len(nodes)])
for idx,node_i in enumerate(nodes):
    for jdx,node_j in enumerate(nodes):
        adjacency_matrix[idx,jdx] = data[node_i][node_j]
        
np.savetxt("adjacency_matrix.csv", adjacency_matrix, delimiter=",")
```

```python
import pandas as pd
pd.DataFrame(adjacency_matrix)
```

### NetworkX

The next cell is the basic code to load the adjacency matrix as a weighted directed network.

```python
import networkx as nx

G = nx.from_numpy_matrix(adjacency_matrix, create_using=nx.DiGraph)
```

Now, we set attributes to the nodes corresponding to their country and industry in order to have them identified

```python
countries_dict = dict(enumerate([i.split()[0] for i in nodes]))
sectors_dict = dict(enumerate([i.split()[1] for i in nodes]))
nx.set_node_attributes(G, countries_dict,'country')
nx.set_node_attributes(G, sectors_dict,'sector')
```

We want to know what type of network we are working with, that is why we want to get basic information about it, like diameter, transitivity, etc.
However, since it is an open network (not all nodes are connected), we cannot use the methods included in NetworkX library. Instead, we will do it semi-manually

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')
degree_sequence = sorted((d for n, d in G_new.degree()), reverse=True)
plt.figure(figsize=(8,8))
plt.plot(degree_sequence,'b-',marker='o')
plt.title("Degree rank plot")
plt.ylabel("degree")
plt.xlabel("rank")
plt.savefig('degree_rank_plot.png')
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')
in_degree_sequence = sorted((d for n, d in G_new.in_degree()), reverse=True)
out_degree_sequence = sorted((d for n, d in G_new.out_degree()), reverse=True)
plt.figure(figsize=(8,8))
plt.plot(in_degree_sequence,'b-',marker='o', color='red', label = 'In')
plt.plot(out_degree_sequence, 'b-', marker = 'o', color='green', label = 'Out')
plt.title("In/Out Degree rank plot")
plt.ylabel("degree")
plt.xlabel("rank")
plt.legend()
plt.savefig('in_out_degree_rank_plot.png')
plt.show()
```

We also want to know how is the distribution of the balance of payments (stregth). For that, we calculate in_strenght - out_strength and plot it sorted

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')
liquidity_dict = {}
for i, j in zip(G_new.in_degree(weight='weight'), G_new.out_degree(weight='weight')):
    liquidity_dict[i[0]] = i[1] - j[1]
liquidity_sorted = sorted((d for n, d in liquidity_dict.items()), reverse=True)
# there are 3 outliers that make the plot look terrible, so we get rid of them
# they are "CHN F", "USA 084" and "USA Q"
liquidity_sorted = liquidity_sorted[:-3]
plt.figure(figsize=(8,8))
plt.plot(liquidity_sorted,'b-',marker='o', color='orange')
plt.title("Balance rank plot")
plt.ylabel("liquidity")
plt.xlabel("rank")
plt.savefig('balance_rank_plot.png')
plt.show()
```

Since there are pairs of nodes that are not connected, we will look for the longest shortest path between any two nodes that are connected. However, we also want to know what proportion of node pairs are not connected over the total amount of node pairs.

```python
shortest_paths = []
for i in G_new.nodes:
    for j in [k for k in G_new.nodes if k != i]:
        try:
            shortest_paths.append((nx.shortest_path_length(G_new, i, j), [i,j]))
        except nx.NetworkXNoPath:
            shortest_paths.append((np.nan, [i,j]))
```

```python
max(shortest_paths)
```

```python
len([i[0] for i in shortest_paths if np.isnan(i[0])]) / len(shortest_paths)
```

```python
# plotting network function
def plot_network(G_n):
    pos = nx.spring_layout(G_n)
    plt.figure(figsize=(12,12))
    plt.axis('off')
    nx.draw_networkx_nodes(G_n, pos)
    nx.draw_networkx_edges(G_n, pos, alpha=0.4)
    nx.draw_networkx_labels(G_n, pos, font_size=12, font_family='verdana')

    plt.show()
```
