Now we create an adjacency matrix to create the financial network to work with.

Rows and columns correspond to the input and output country-industry respectively.

```python
import numpy as np

nodes = [c + " " + i for c in country.keys() for i in industry.keys()]
data = data.loc[nodes][nodes]

adjacency_matrix = np.zeros([len(nodes),len(nodes)])
for idx,node_i in enumerate(nodes):
    for jdx,node_j in enumerate(nodes):
        adjacency_matrix[idx,jdx] = data[node_i][node_j]
        
np.savetxt("adjacency_matrix.csv", adjacency_matrix, delimiter=",")
```

```python
import pandas as pd
pd.DataFrame(adjacency_matrix)
```

### NetworkX

The next cell is the basic code to load the adjacency matrix as a weighted directed network.

```python
import networkx as nx

G = nx.from_numpy_matrix(adjacency_matrix, create_using=nx.MultiDiGraph)
```

Now, we set attributes to the nodes corresponding to their country and industry in order to have them identified

```python
countries_dict = dict(enumerate([i.split()[0] for i in nodes]))
sectors_dict = dict(enumerate([i.split()[1] for i in nodes]))
nx.set_node_attributes(G, countries_dict,'country')
nx.set_node_attributes(G, sectors_dict,'sector')
```

Computing the betweenness and ranking the most important nodes according to it 

```python
# calculating the betweenness 
import networkx as nx 
b=nx.betweenness_centrality(G) 
```

```python
# setting a threshold of 0.0004, grouping the betweeness with the country/industry and sorting it
test_1=[]
test_2=[]
for i in b:
    if b[i]>=0.0004:
        test_3=i
        test_1.append(test_3)
        test_3=b[i]
        test_2.append(test_3)
top_betweenness= pd.DataFrame(test_2,test_1)
top_betweenness.columns=['Betweenness']
top_betweenness=top_betweenness.sort_values(['Betweenness'], ascending=0)
top_betweenness['Index']=top_betweenness.index
top_betweenness.reset_index(drop=True, inplace=True)
j=0
top_betweenness['Name']=0
for i in top_betweenness['Index']:
    top_betweenness['Name'][j]=data.index[i]
    j=j+1
top_betweenness.drop(['Index'],axis=1)[:10]
```

First, let's look at the distribution of the input-output relationships (or transactions)

```python
edges = []
for i in G.nodes:
    for j in [k for k in G.nodes if k != i]:
        edge = G.get_edge_data(i,j,0)
        if edge is not None:
            edges.append(edge['weight'])
```

We want to know what type of network we are working with, that is why we want to get basic information about it, like diameter, transitivity, etc.
However, since it is an open network (not all nodes are connected), we cannot use the methods included in NetworkX library. Instead, we will do it semi-manually

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')
degree_sequence = sorted((d for n, d in G.degree()), reverse=True)
plt.figure(figsize=(8,8))
plt.plot(degree_sequence,'b-',marker='o')
plt.title("Degree rank plot")
plt.ylabel("degree")
plt.xlabel("rank")
plt.savefig('degree_rank_plot.png')
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')
in_degree_sequence = sorted((d for n, d in G.in_degree()), reverse=True)
out_degree_sequence = sorted((d for n, d in G.out_degree()), reverse=True)
plt.figure(figsize=(8,8))
plt.plot(in_degree_sequence,'b-',marker='o', color='red', label = 'In')
plt.plot(out_degree_sequence, 'b-', marker = 'o', color='green', label = 'Out')
plt.title("In/Out Degree rank plot")
plt.ylabel("degree")
plt.xlabel("rank")
plt.legend()
plt.savefig('in_out_degree_rank_plot.png')
plt.show()
```

We also want to know how is the distribution of the balance of payments (stregth). For that, we calculate in_strenght - out_strength and plot it sorted

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')
liquidity_dict = {}
for i, j in zip(G.in_degree(weight='weight'), G.out_degree(weight='weight')):
    liquidity_dict[i[0]] = i[1] - j[1]
liquidity_sorted = sorted((d for n, d in liquidity_dict.items()), reverse=True)
# there are 3 outliers that make the plot look terrible, so we get rid of them
# they are "CHN F", "USA 084" and "USA Q"
liquidity_sorted = liquidity_sorted[:-3]
plt.figure(figsize=(8,8))
plt.plot(liquidity_sorted,'b-',marker='o', color='orange')
plt.title("Balance rank plot")
plt.ylabel("liquidity")
plt.xlabel("rank")
plt.savefig('balance_rank_plot.png')
plt.show()
```

Since there are pairs of nodes that are not connected, we will look for the longest shortest path between any two nodes that are connected. However, we also want to know what proportion of node pairs are not connected over the total amount of node pairs.

```python
shortest_paths = []
for i in G.nodes:
    for j in [k for k in G.nodes if k != i]:
        try:
            shortest_paths.append((nx.shortest_path_length(G, i, j), [i,j]))
        except nx.NetworkXNoPath:
            shortest_paths.append((np.nan, [i,j]))
```

```python
max(shortest_paths)
```

```python
len([i[0] for i in shortest_paths if np.isnan(i[0])]) / len(shortest_paths)
```

We see that the longest distance between any 2 given nodes is 3 edges, but 12.96% of the possible pairs of nodes are not connected. Now, we will look at transitivity, connectivity and clustering coefficient to check the interconnection and dependence of the network.
Since we have a directed network with self-loops but no paralell edges, we transform it into a DiGraph in order to find these metrics.

```python
G_di = nx.DiGraph(G)
```

```python
nx.average_clustering(G_di)
```

```python
nx.transitivity(G_di)
```

## Feature matrices creation

We aggregate all inputs by country

```python
inputs_data = pd.DataFrame(index=data.index)
        
for c in country:
    cols = [c + " " + i for i in industry]
    inputs_data[c] = data[cols].sum(axis=1)
```

We aggregate all outputs by country

```python
outputs_data = pd.DataFrame(index=data.index)
        
for c in country:
    rows = [c + " " + i for i in industry]
    outputs_data[c] = data.loc[rows].sum()
```

We create a table to express country input and output relationships of country-industry pairs

```python
country_industry = pd.concat([inputs_data.add_prefix("I "), outputs_data.add_prefix("O ")], axis=1)
country_industry.to_csv("country_industry.csv")
```
