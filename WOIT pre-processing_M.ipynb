Now we create an adjacency matrix to create the financial network to work with.

Rows and columns correspond to the input and output country-industry respectively.

```python
import numpy as np

nodes = [c + " " + i for c in country.keys() for i in industry.keys()]
data = data.loc[nodes][nodes]

adjacency_matrix = np.zeros([len(nodes),len(nodes)])
for idx,node_i in enumerate(nodes):
    for jdx,node_j in enumerate(nodes):
        adjacency_matrix[idx,jdx] = data[node_i][node_j]
        
np.savetxt("adjacency_matrix.csv", adjacency_matrix, delimiter=",")
```

### NetworkX

The next cell is the basic code to load the adjacency matrix as a weighted directed network.

```python
import networkx as nx

G = nx.from_numpy_matrix(adjacency_matrix, create_using=nx.MultiDiGraph)
```

```python
countries_dict = dict(enumerate([i.split()[0] for i in nodes]))
sectors_dict = dict(enumerate([i.split()[1] for i in nodes]))
nx.set_node_attributes(G, countries_dict,'country')
nx.set_node_attributes(G, sectors_dict,'sector')
```

## Feature matrices creation

We aggregate all inputs by country

```python
inputs_data = pd.DataFrame(index=data.index)
        
for c in country:
    cols = [c + " " + i for i in industry]
    inputs_data[c] = data[cols].sum(axis=1)
```

We aggregate all outputs by country

```python
outputs_data = pd.DataFrame(index=data.index)
        
for c in country:
    rows = [c + " " + i for i in industry]
    outputs_data[c] = data.loc[rows].sum()
```

We create a table to express country input and output relationships of country-industry pairs

```python
country_industry = pd.concat([inputs_data.add_prefix("I "), outputs_data.add_prefix("O ")], axis=1)
country_industry.to_csv("country_industry.csv")
```

```python
def plot_network(G):
    pos = nx.spring_layout(G)
    plt.figure(figsize=(12,12))
    plt.axis('off')
    nx.draw_networkx_nodes(G, pos)
    nx.draw_networkx_edges(G, pos, alpha=0.4)
    nx.draw_networkx_labels(G, pos, font_size=12, font_family='verdana')

    plt.show()
```
